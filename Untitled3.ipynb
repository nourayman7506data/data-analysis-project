```python
#Classification
# Load the data
df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
df['paid_lunch'] = df['lunch'].map(lambda x: '1' if x =='standard'  else '0')
df['preparation'] = df['test preparation course'].map(lambda x: '1' if x =='completed'  else '0')
df['pass_math'] = df['math score'].map(lambda x: '1' if x >50  else '0')
df['pass_reading'] = df['reading score'].map(lambda x: '1' if x >50 else '0')
df['pass_writing'] = df['writing score'].map(lambda x: '1' if x >50  else '0')
df
```

```python
# Drop columns that we won't use in the analysis
df = df.drop(['race/ethnicity','lunch','test preparation course','math score','reading score','writing score'], axis=1)
df
```

```python
# Remove any missing values

df = df.dropna()
df
```

```python
# Convert categorical variables to dummy variables

df = pd.get_dummies(df, columns=['gender', 'parental level of education'])
df
```

```python
# Split the data into training and testing sets

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('pass_math', axis=1), df['pass_math'], test_size=0.2, random_state=0)
```

```python
# Fit a logistic regression model

from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
# Make predictions on the testing set

y_pred = log_reg.predict(X_test)
```

```python
# Evaluate the accuracy of the model

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```
